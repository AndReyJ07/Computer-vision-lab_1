**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB) based on https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html**

```python
image = cv2.imread("./Lab_1/giraffes.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
kernel = np.ones((4,4),np.float32)/10
```

```python
dst = cv2.filter2D(image,-1,kernel)
```

```python
plt.imshow(dst)
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image based on https://docs.opencv.org/4.x/d1/db7/tutorial_py_histogram_begins.html. Pay careful attention to the ordering of the channels.**

```python
image = cv2.imread("./Lab_1/giraffes.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
redHist = cv2.calcHist([image], [2], None, [256], [0, 256])
greenHist = cv2.calcHist([image], [1], None, [256], [0, 256])
blueHist = cv2.calcHist([image], [0], None, [256], [0, 256])

plt.plot(redHist, color="r")
plt.plot(greenHist, color="g")
plt.plot(blueHist, color="b")
```
