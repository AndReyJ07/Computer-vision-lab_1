#### TASK: Open the *dog_backpack.jpg* image from any folder and display it in the notebook. Make sure to correct for the RGB order. The image can be pre-downloaded from Internet using wget command 

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from PIL import Image
```

```python
image = Image.open('./Lab_1/index.png')
```

```python
plt.imshow(image)
```

```python
npImage = np.array(image)
correctImg = cv2.cvtColor(npImage, cv2.COLOR_BGR2RGB)
image = Image.fromarray(correctImg.astype(np.uint8))
plt.imshow(image)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
image = cv2.imread('./Lab_1/index.png')
pts = np.array([[110, 55], [70, 120], [150, 120]], np.int32)

isClosed = True
color = (0, 0, 255)
thickness = 2

image = cv2.polylines(image, [pts], isClosed, color, thickness)

plt.imshow(image)
```
